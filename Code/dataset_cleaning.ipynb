```python
# dependencies
import pandas as pd
```

```python
# create file path
path = "../Resources/Crime_Data_from_2020_to_Present.csv"

# read file to DataFrame
orig_df = pd.read_csv(path)
```

```python
# show DataFrame
cleaning_df = orig_df
orig_df.head()
```

```python
# delete unnecessary columns
del cleaning_df['DR_NO']
del cleaning_df['AREA']
del cleaning_df['Rpt Dist No']
del cleaning_df['Part 1-2']
del cleaning_df['Mocodes']
del cleaning_df['Premis Cd']
del cleaning_df['Weapon Used Cd']
del cleaning_df['Weapon Desc']
del cleaning_df['Status']
del cleaning_df['Status Desc']
del cleaning_df['Crm Cd 1']
del cleaning_df['Crm Cd 2']
del cleaning_df['Crm Cd 3']
del cleaning_df['Crm Cd 4']
del cleaning_df['LOCATION']
del cleaning_df['Cross Street']

cleaning_df.head()
```

```python
# rename columns
cleaning_df = cleaning_df.rename(columns={
    "Date Rptd": "Date Reported",
    "DATE OCC": "Date Occurred",
    "TIME OCC": "Time Occurred",
    "AREA NAME": "Geographic Area",
    "Crm Cd": "Crime Code",
    "Crm Cd Desc": "Crime Code Description",
    "Vict Age": "Victim Age",
    "Vict Sex":	"Victim Sex",
    "Vict Descent":	"Victim Ethnicity",
    "Premis Desc": "Premise Description",
    "LAT": "Latitude",
    "LON": "Longitude"
})

cleaning_df.head()
```

```python
cleaned_df = cleaning_df
cleaned_df.to_csv("../Resources/Cleaned_Data.csv")
```
