## Victim Age Groups

```python
age_bins = [0, 18, 25, 45, 65, 120]
labels = ["<18", "18-25", "26-45", "46-65", ">65"]

crime_data_copy["Victim Age Range"] = pd.cut(crime_data_copy["Victim Age"], age_bins, labels=labels, include_lowest=True)
```
